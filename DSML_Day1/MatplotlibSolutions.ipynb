__Line Plot__<br><br>

The CSV file sales_data.csv contains the following data:<br>
<img src="nbimages/sales_data.png" width=600 height=500/>

Use the CSV file to construct a Panda DataFrame.<br>

Using the DataFrame compute the total profit of each month and show it using a line plot.<br>

The line plot should look like this.<br><br>

<img src="nbimages/lineplot_1.png" width=400 height=400/>

```python
import pandas as pd
import matplotlib.pyplot as plt  
import datetime

plt.style.use('default')
df = pd.read_csv("datasets/sales_data.csv")

fig=plt.figure()
ax=fig.add_subplot(111)


profitList = df ['total_profit']
monthList  = df ['month_number']

ax.plot(monthList, profitList, label = 'Month-wise Profit data of last year')

ax.set_xlabel('Month number')
ax.set_ylabel('Profit in dollar')
ax.set_xticks(monthList)
ax.set_title('Company profit per month')
ax.set_yticks(range(100000, 600000, 100000))
plt.tight_layout()
plt.savefig("lineplot_1.png",dpi=300, bbox_inches = "tight") # png, svg, pdf
plt.show();
```

__Line Plot__ 2<br><br>

Update the line plot to plot, on the X axis, the month name instead of the month number.<br>

The line plot should look like this.<br><br>

<img src="nbimages/lineplot_2.png" width=400 height=400/>

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
profitList = df ['total_profit']
monthList  = df ['month_number']
ax.plot(monthList, profitList, label = 'Month-wise Profit data of last year')
ax.set_xlabel('Month')
ax.set_ylabel('Profit in dollar')
ticklabels = [datetime.date(1900, item, 1).strftime('%b') for item in monthList]
ax.set_xticks(monthList)
ax.set_xticklabels(ticklabels)
ax.set_title('Company profit per month')
ax.set_yticks(range(100000, 600000, 100000))
plt.tight_layout()
plt.show();
```

__Line Plot__ 3<br><br>

Update the prevous line plot and use the following Style properties:<br>

* Line should be a "dotted line" and the line color should be red.<br>
* Show legend at the lower right location.<br>
* Add a circle marker.<br>
* Line marker color should be red.<br>
* Line width should be 3.<br>

The line plot should look like this.<br><br>

<img src="nbimages/lineplot_3.png" width=400 height=400/>

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
profitList = df ['total_profit']
monthList  = df ['month_number']
ax.plot(monthList, profitList, label = 'Month-wise Profit data of last year',color='r', marker='o', markerfacecolor='k', 
      linestyle='--', linewidth=3)
ax.set_xlabel('Month')
ax.set_ylabel('Profit in dollar')
ticklabels = [datetime.date(1900, item, 1).strftime('%b') for item in monthList]
ax.set_xticks(monthList)
ax.set_xticklabels(ticklabels)
ax.set_title('Company profit per month')
ax.set_yticks(range(100000, 600000, 100000))
plt.legend(loc='lower right')
plt.savefig("lineplot_3.png",dpi=300, bbox_inches = "tight")
```

__Scatter Plot__<br><br>

Plot the toothpaste sales data of each month using a scatter plot.<br>
Add a grid in the plot with dashed lines.

The scatter plot should look like this.<br><br>

<img src="nbimages/scatterplot.png" width=400 height=400/>


```python
plt.style.use('default') # seaborn bmh classic ggplot ....
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
monthList  = df ['month_number']
toothPasteSalesData = df ['toothpaste']

ax.scatter(monthList, toothPasteSalesData, label = 'Tooth paste Sales data')
ax.set_xlabel('Month number')
ticklabels = [datetime.date(1900, item, 1).strftime('%b') for item in monthList]
ax.set_xlim(1,12)
ax.set_xticks(monthList)
ax.set_xticklabels(ticklabels)
ax.set_title('Tooth paste Sales data')
ax.legend(loc='best')
ax.grid(True, linewidth= 1, linestyle="--");
#plt.show()
```

__Bar Chart__ 1<br><br>

Plot the face cream and the face wash product sales using a bar chart.<br>
The 2 bar charts should be displayed one on top of the other. <br>

The Bar chart should look like this.<br><br>

<img src="nbimages/barchart1.png" width=400 height=400/>

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

monthList  = df ['month_number']
faceCreamSalesData   = df ['facecream']
faceWashSalesData   = df ['facewash']

ax.bar(monthList, faceCreamSalesData, label = 'Face Cream sales data', align='edge')
ax.bar(monthList, faceWashSalesData, label = 'Face Wash sales data', align='edge')
ax.set_xlabel('Month number')
ax.set_ylabel('Sales units in number')

ticklabels = [datetime.date(1900, item, 1).strftime('%b') for item in monthList]
ax.set_xticks(monthList)
ax.set_xticklabels(ticklabels)
ax.set_title('Facewash and Facecream Sales data')

ax.legend(loc='best')
ax.grid(True, linewidth= 1, linestyle="--");
#plt.show()
```

__Bar Chart__ 2<br><br>

Plot the face cream and the face wash product sales using a bar chart.<br>
Two separate adjacent bar charts should now be used to display the number of units sold per month for each product. <br>

The Bar chart should look like this.<br><br>

<img src="nbimages/barchart2.png" width=400 height=400/>

```python
monthList  = df ['month_number']

ax=df.plot(x="month_number", y=['facecream','facewash'], kind="bar")
ax.set_xlabel('Month number')
ax.set_ylabel('Sales units in number')
ticklabels = [datetime.date(1900, item, 1).strftime('%b') for item in monthList]
ax.set_xticks(monthList.index)
ax.set_xticklabels(ticklabels)
ax.set_title('Facewash and Facecream Sales data')
ax.legend(loc='best');
#plt.show();
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

monthList  = df ['month_number']
faceCreamSalesData   = df ['facecream']
faceWashSalesData   = df ['facewash']

ax.bar(monthList, faceCreamSalesData, width=0.25, label = 'Face Cream sales data', align='edge')
ax.bar(monthList+0.25, faceWashSalesData, width=0.25,label = 'Face Wash sales data', align='edge')
ax.set_xlabel('Month number')
ax.set_ylabel('Sales units in number')

ticklabels = [datetime.date(1900, item, 1).strftime('%b') for item in monthList]
ax.set_xticks(monthList+0.25)
ax.set_xticklabels(ticklabels)
ax.set_title('Facewash and Facecream Sales data')
ax.legend(loc='best')
ax.grid(True, linewidth=1, linestyle="--");
#plt.show();
```

__Pie Chart__<br><br>
Calculate total sale data for each product and show it using a _Pie chart_.<br>
Note: In the Pie chart display, for each product, the percentage of units sold per year.<br>

The Pie chart should look like this.<br><br>

<img src="nbimages/piechart.png" width=500 height=500/>

```python
plt.style.use('default') # seaborn bmh classic ggplot ....

fig=plt.figure()
ax=fig.add_subplot(1,1,1)

monthList  = df ['month_number']
ax.axis("equal")
labels=df.columns[1:7]
ax.pie(df.loc[:,'facecream':'moisturizer'].sum(), labels=labels, autopct='%1.1f%%')
ax.legend(bbox_to_anchor=(-0.3, -0.7, 0.5, 1.5))
ax.set_title('Sales data');
# plt.show(); # Only needed outside of a Jupyter notebook
```

__Stack Plot__<br><br>
Read all product sales data and show it using a __stack plot__.<br>
The Stack plot should look like this.<br><br>
<img src="nbimages/stackplot.png" width=400 height=400/>

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

monthList  = df ['month_number']

ax.stackplot(monthList, df.loc[:,'facecream':'moisturizer'].T, labels=df.columns[1:7])

plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
ax.legend(loc='best')
ax.set_title('Sales data');
#plt.show()
```
